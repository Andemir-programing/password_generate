```python
import string
import random
```

```python
# генерптор паролей
def pass_gen(length):
  # необходимые символы
  digits = string.digits
  punct = string.punctuation
  letters_u = string.ascii_uppercase
  letters_l = string.ascii_lowercase
  # переменная для пароля
  password = ''
  var = [digits, letters_u, letters_l, punct]
  # проверка длины пароля
  if length < 12:
    return 'Ошибка! Пароль должен содержать не менее 12 символов'

  # генерация пароля
  else:
    password += random.choice(digits)
    password += random.choice(letters_u)  
    password += random.choice(letters_l)
    password += random.choice(punct)    

    while len(password) < length:
      password += random.choice(var[random.randint(0,3)])
      return password 
```

```python
sites = ['VK', 'Steam', 'Facebook', 'MyBook', 'Spotify']
passwords = []
for _ in range(len(sites)):
  passwords.append(pass_gen(15))
```
